```python
def read_cluster_file(cluster_file):
    """Lit un fichier de clustering de séquences et renvoie un dictionnaire de clusters."""
    clusters = {}
    with open(cluster_file, 'r') as f:
        for line in f:
            seq, cluster = line.strip().split()
            if cluster not in clusters:
                clusters[cluster] = []
            clusters[cluster].append(seq)
    return clusters

def count_sequences(clusters):
    """Compte le nombre de séquences dans chaque cluster."""
    counts = {}
    for cluster, seqs in clusters.items():
        counts[cluster] = len(seqs)
    return counts

# Exemple d'utilisation
cluster_file = '/Users/anis/clu.tsv'
clusters = read_cluster_file(cluster_file)
counts = count_sequences(clusters)
for cluster, count in counts.items():
    print(f'Cluster {cluster}: {count} sequences')
```
