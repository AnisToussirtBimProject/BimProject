```python
import os
from Bio.PDB import PDBParser

def get_pdb_chains(dossier):
    """Retourne une liste de toutes les chaînes protéiques des fichiers PDB dans le dossier spécifié."""
    chains = []
    for filename in os.listdir(dossier):
        if filename.endswith('.pdb'):
            chemin_fichier = os.path.join(dossier, filename)
            parser = PDBParser(QUIET=True)
            structure = parser.get_structure(filename, chemin_fichier)
            for model in structure:
                for chain in model:
                    if chain.get_id() not in chains:
                        chains.append(chain.get_id())
    return chains

d = '/Users/anis/Downloads/ALL-experimental'
print(get_pdb_chains(d))
```

```python
import os

# Chemin du fichier texte
chemin_fichier = "/Users/anis/Downloads/msa_output_2/OUTPUT_KINASES/0a3m"

# Dossier contenant les fichiers PDB
dossier_pdb = "/Users/anis/Downloads/db_all/pdb"

# Lecture du contenu du fichier texte
with open(chemin_fichier, "r") as f:
    contenu = f.readline().strip().split("\t")


# Nom du fichier PDB
nom_pdb = contenu[0].strip(">.pdb")

# Séquence à extraire
sequence = contenu[1]


# Chemin vers le fichier PDB
chemin_pdb = os.path.join(dossier_pdb, nom_pdb + ".pdb")

# Extraction de la séquence de chaque chaîne
with open(chemin_pdb, "r") as f:
    lignes = f.readlines()

# Dictionnaire pour stocker les séquences extraites
sequences = {}

# Itération sur les lignes du fichier PDB
for ligne in lignes:
    # Vérification si la ligne contient une séquence
    if ligne.startswith("ATOM") or ligne.startswith("HETATM"):
        chaine = ligne[21]
        if chaine not in sequences:
            sequences[chaine] = ""
        residu = ligne[17:20]
        # Vérification si le résidu est un acide aminé
        if residu.strip() in ["ALA", "ARG", "ASN", "ASP", "CYS", "GLN", "GLU", "GLY", "HIS", "ILE",
                              "LEU", "LYS", "MET", "PHE", "PRO", "SER", "THR", "TRP", "TYR", "VAL"]:
            resseq = ligne[22:26]
            # Vérification si le résidu est présent dans la séquence à extraire
            if resseq.strip() in sequence:
                sequences[chaine] += ligne[17:20]

# Écriture des séquences extraites dans des fichiers PDB séparés
for chaine, seq in sequences.items():
    with open(nom_pdb + chaine + ".pdb", "w") as f:
        f.write(f"REMARK  Extracted from {nom_pdb}.pdb\n")
        f.write(f"REMARK  Sequence: {seq}\n")
        for ligne in lignes:
            if ligne[21] == chaine:
                f.write(ligne)
```

```python
import os

# Définition d'une fonction pour extraire les séquences de protéines à partir du fichier texte
def extract_sequences(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
        sequences = {}
        current_key = None
        current_sequence = ""
        for line in lines:
            if line.startswith(">"):
                if current_key is not None:
                    sequences[current_key] = current_sequence
                current_key = line.strip()[1:]
                current_sequence = ""
            else:
                current_sequence += line.strip()
        if current_key is not None:
            sequences[current_key] = current_sequence
    return sequences

# Définition d'une fonction pour modifier les fichiers PDB en ne gardant que la partie de la séquence correspondante
def modify_pdb_files(sequences, pdb_dir):
    for pdb_file in os.listdir(pdb_dir):
        if pdb_file.endswith(".pdb"):
            pdb_path = os.path.join(pdb_dir, pdb_file)
            with open(pdb_path, 'r') as f:
                lines = f.readlines()
                new_lines = []
                for line in lines:
                    if line.startswith("ATOM"):
                        residue_num = line[22:26].strip()
     
                        chain_id = line[21]
                        pdb_key = "{}{}".format(chain_id, pdb_file.split(".")[0])
                        if pdb_key in sequences:
                            sequence = sequences[pdb_key]
                            if residue_num <= len(sequence):
                                new_line = line[:16] + sequence[residue_num-1] + line[17:]
                                new_lines.append(new_line)
                with open(pdb_path, 'w') as f:
                    f.writelines(new_lines)

# Exemple d'utilisation
if __name__ == "__main__":
    sequences = extract_sequences("/Users/anis/Downloads/msa_output_2/0a3m copie")
    pdb_dir = "/Users/anis/Downloads/db_all/pdb"
    modify_pdb_files(sequences, pdb_dir)
```
