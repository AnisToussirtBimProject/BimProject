```python
########

def count_pdb_files(filename):
    pdb_set = set()
    with open(filename, 'r') as file:
        for line in file:
            pdb_files = line.split(',')
            for pdb_file in pdb_files:
                pdb_name = pdb_file.strip().split(':')[0].strip()
                if pdb_name.endswith('.pdb'):
                    pdb_set.add(pdb_name)
    return len(pdb_set)

print(count_pdb_files("/Users/anis/Downloads/msa_output/TAILLES_PDB"))
```

```python
def compare_files(file1, file2):
    with open(file1) as f1, open(file2) as f2:
        lines1 = [line.strip() for line in f1]
        lines2 = [line.strip() for line in f2]
        
    numbers_dict1 = {}
    for line in lines1:
        name, num = line.split(':')
        numbers_dict1[name] = int(num)
        
    numbers_dict2 = {}
    for line in lines2:
        name, num = line.split(' : ')
        numbers_dict2[name] = int(num)
    
    comparison_list = []
    for name in numbers_dict1.keys():
        if name in numbers_dict2:
            comparison_list.append((name, numbers_dict1[name], numbers_dict2[name]))
    
    return comparison_list

comparison_list = compare_files("/Users/anis/Downloads/msa_output/TAILLES_PDB_bon", "/Users/anis/Downloads/msa_output/TAILLES_PDB_2")
print(comparison_list)
```

```python
def check_duplicates(file_path):
    file_set = set()
    with open(file_path, 'r') as f:
        for line in f:
            file_name = line.strip().split()[0]
            if file_name in file_set:
                print(f"Le fichier {file_name} est en double.")
            else:
                file_set.add(file_name)
    print("Vérification terminée.")

check_duplicates("/Users/anis/Downloads/msa_output/COMPARAISON_TAILLES")
```

```python
def count_chains(file_path):
    chain_counts = {}
    with open(file_path, 'r') as f:
        for line in f:
            file_name = line.strip().split()[0]
            pdb_code = file_name[:4]
            chain = file_name[4].upper()
            if pdb_code in chain_counts:
                chain_counts[pdb_code].add(chain)
            else:
                chain_counts[pdb_code] = set(chain)
    for pdb_code, chains in chain_counts.items():
        print(f"Le fichier PDB {pdb_code} contient {len(chains)} chaîne(s).")

count_chains("/Users/anis/Downloads/msa_output/COMPARAISON_TAILLES")
```

```python
def count_chains(file_path):
    chain_counts = {}
    with open(file_path, 'r') as f:
        for line in f:
            file_name = line.strip().split()[0]
            print(file_name)
            pdb_code = file_name[:4]
            chain = file_name[4].upper()
            if pdb_code in chain_counts:
                chain_counts[pdb_code].add(chain)
            else:
                chain_counts[pdb_code] = set(chain)
    sorted_chains = sorted(chain_counts.items(), key=lambda x: len(x[1]), reverse=True)
    for pdb_code, chains in sorted_chains:
        print(f"{pdb_code} : {len(chains)} chaîne(s).")

count_chains("/Users/anis/Downloads/msa_output/COMPARAISON_TAILLES")
```

```python
import os
from ase.io import read

# Chemin d'accès au dossier contenant les fichiers CIF
dossier = '/Users/anis/Desktop/ALL_CIF'

# Liste des noms de fichiers sans l'extension .cif
noms_fichiers = []
for nom_fichier in os.listdir(dossier):
    if nom_fichier.endswith('.cif'):
        nom_sans_ext = os.path.splitext(nom_fichier)[0]
        noms_fichiers.append(nom_sans_ext)

# Liste des chaînes dans chaque fichier
chaines_par_fichier = {}
for nom_fichier in noms_fichiers:
    fichier_cif = os.path.join(dossier, nom_fichier + '.cif')
    atome = read(fichier_cif)
    print(atome)
    chaines_par_fichier[nom_fichier] = atome.get_chemical_symbols()

# Afficher les chaînes dans chaque fichier
for nom_fichier, chaines in chaines_par_fichier.items():
    print(f"Les chaînes dans le fichier {nom_fichier} sont : {chaines}")
```
